```python
# define features
X = ml_df.drop(columns='Item_Outlet_Sales')
#define target
y=ml_df['Item_Outlet_Sales']
```

```python
# train split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
#scaler
scaler = StandardScaler()
#One-hot encoder
ohe = OneHotEncoder(handle_unknown='ignore',sparse=False)
#Imputer
#freqent_imputer = SimpleImputer(strategy='most_frequent')
median_imputer = SimpleImputer(strategy='median')
constant_imputer = SimpleImputer(strategy='constant',fill_value='missing')

```

```python
#instantiate pipeline
number_pipe = make_pipeline(median_imputer,scaler)
nominal_pipe = make_pipeline(constant_imputer,ohe)
```

```python
df.columns
```

```python
df.info()
```

```python
# instantiate selector
number_selector = make_column_selector(dtype_include='number')
nominal_cols = ['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Establishment_Year','Outlet_Location_Type','Outlet_Size','Outlet_Type']
#ordinal_cols = ['Outlet_Size']
# tuple for column transformer
number_tuple = (number_pipe,number_selector)
#ordinal_tuple = (ordinal_pipe,ordinal_cols)
nominal_tuple = (nominal_pipe,nominal_cols)
#instantiate transformers
preprocessor = make_column_transformer(nominal_tuple,number_tuple,remainder='drop')
preprocessor
```

```python
#fit on train
preprocessor.fit(X_train)
```

```python
# transform train and test
X_train_preprocessed = preprocessor.transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

```python
# Check for missing values and that data is scaled and one-hot encoded
print(np.isnan(X_train_preprocessed).sum().sum(), 'missing values in training data')
print(np.isnan(X_test_preprocessed).sum().sum(), 'missing values in testing data')
print('\n')
print('All data in X_train_processed are', X_train_preprocessed.dtype)
print('All data in X_test_processed are', X_test_preprocessed.dtype)
print('\n')
print('shape of data is', X_train_preprocessed.shape)
print('\n')
X_train_preprocessed
```
